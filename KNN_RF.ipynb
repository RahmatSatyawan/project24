Convert 'AnnualRevenue' and 'LatePayment' from USD to IDR

```python
# 15735 at 20-03-2024
kurs_usd_to_idr = 15735

# convert to IDR
df_smes['AnnualRevenueIDR'] = df_smes['AnnualRevenue'] * kurs_usd_to_idr
df_smes['LatePaymentsIDR'] = df_smes['LatePayments'] * kurs_usd_to_idr

# drop column
df_smes.drop(['LatePayments', 'AnnualRevenue'], axis=1, inplace=True)

print(df_smes)
```

Convert 'CreditScore' from 300-850 to 250-900 (indonesia)

```python
# Define the conversion function From FICO Score
def convert_to_category(score):
    if score >= 740:
        return "Excellent"
    elif 700 <= score <= 739:
        return "Very Good"
    elif 670 <= score <= 699:
        return "Good"
    elif 580 <= score <= 669:
        return "Fair"
    else:
        return "Poor"

# Apply the conversion function to create a new column 'CreditCategory'
df_smes['CreditCategory'] = df_smes['CreditScore'].apply(convert_to_category)


# Drop column
df_smes.drop(['CreditScore'], axis=1, inplace=True)
print(df_smes)
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Define your custom mapping of categories to numerical labels
custom_mapping = {
    'Excellent' : 1,
    'Very Good': 2,
    'Good': 3,
    'Fair': 4,
    'Poor': 5
}

# Fit the label encoder with your custom mapping
df_smes['CreditCategoryEnc'] = df_smes['CreditCategory'].map(custom_mapping)

# Display unique categories and their corresponding encoded labels
print("Unique Categories:", df_smes['CreditCategory'].unique())
print("Encoded Labels:", df_smes['CreditCategoryEnc'].unique())
```

```python
#Z-Score
print(f'Before filter outlier: {len(df_smes)}')

filtered_entries = np.array([True] * len(df_smes))
for col in [ 'Industry_Healthcare', 'Industry_Retail', 'Industry_Technology',
            'AnnualRevenueIDR', 'LatePaymentsIDR', 'CreditCategoryEnc'
            ]:
    zscore = abs(stats.zscore(df_smes[col]))
    filtered_entries = (zscore < 3) & filtered_entries

df_smes = df_smes[filtered_entries]

print(f'After filter outlier: {len(df_smes)}')
```

```python
# Drop Target Value 'CreditScoreConv'
predictor_features = ['Industry_Healthcare', 'Industry_Retail', 'Industry_Technology', 'AnnualRevenueIDR', 'LatePaymentsIDR']

X = df_smes[predictor_features]
y = df_smes['CreditCategory']
```

```python
# Split Train - Test (Test size 20% rd1)
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)

y_pred_knn = knn_model.predict(X_val)
print(classification_report(y_val,y_pred_knn))
print(confusion_matrix(y_val,y_pred_knn))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_val)
print(classification_report(y_val,y_pred_rf))
print(confusion_matrix(y_val,y_pred_rf))
```

```python
# Classification reports
rf_report = classification_report(y_val, y_pred_rf, output_dict=True)
knn_report = classification_report(y_val, y_pred_knn, output_dict=True)

# Extracting precision, recall, and F1-score for each class
classes = list(rf_report.keys())[:-3]
metrics = ['precision', 'recall', 'f1-score']

rf_scores = np.array([[rf_report[label][metric] for metric in metrics] for label in classes])
knn_scores = np.array([[knn_report[label][metric] for metric in metrics] for label in classes])

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(8, 10), sharex=True)

for i, metric in enumerate(metrics):
    ax[i].bar(np.arange(len(classes)) - 0.2, rf_scores[:, i], width=0.4, label='Random Forest', color='b')
    ax[i].bar(np.arange(len(classes)) + 0.2, knn_scores[:, i], width=0.4, label='K-Nearest Neighbors', color='r')
    ax[i].set_ylabel(metric.capitalize())
    ax[i].set_xticks(np.arange(len(classes)))
    ax[i].set_xticklabels(classes)
    ax[i].legend()

plt.xlabel('Classes')
plt.suptitle('Classification Report Comparison')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Classification Report:
- Precision: Precision measures the proportion of
true positive predictions among all positive predictions. For example, for the 'Poor' class, the precision is 0.50, which means that among all instances predicted as 'Poor', only 50% are actually 'Poor'.
- Recall: Recall (also called sensitivity) measures the proportion of true positive predictions among all actual positive instances. For example, for the 'Excellent' class, the recall is 0.20, which means that only 20% of all 'Excellent' instances were correctly predicted as 'Excellent'.
- F1-score: The F1-score is the harmonic mean of precision and recall. It provides a balance between precision and recall. It is useful when you have an uneven class distribution.
Support: Support is the number of actual occurrences of the class in the validation dataset.

## Confusion Matrix:
A confusion matrix is a table that is often used to describe the performance of a classification model on a set of test data for which the true values are known. It presents a summary of the model's predictions against the actual labels.

Each row of the matrix represents the instances in an actual class.
Each column represents the instances in a predicted class.
The diagonal elements represent the number of correctly classified instances for each class.
Off-diagonal elements represent misclassifications.
