Move Credit Score as Target to last right column

```python
columns = ['AnnualRevenue', 'CreditScore', 'LatePayments', 'Industry_Healthcare', 'Industry_Retail', 'Industry_Technology']
column_to_move = 'CreditScore'

# Perform the move
column = df_smes.pop(column_to_move)
df_smes.insert(len(df_smes.columns), column.name, column)
```

Convert 'AnnualRevenue' and 'LatePayment' from USD to IDR

```python
# 15735 at 20-03-2024
kurs_usd_to_idr = 15735

# convert to IDR
df_smes['AnnualRevenueIDR'] = df_smes['AnnualRevenue'] * kurs_usd_to_idr
df_smes['LatePaymentsIDR'] = df_smes['LatePayments'] * kurs_usd_to_idr

# drop column
df_smes.drop(['LatePayments', 'AnnualRevenue'], axis=1, inplace=True)

print(df_smes)
```

Convert 'CreditScore' from 300-850 to 250-900 (indonesia)

```python
df_smes.describe()
```

```python
# Define the conversion function
def convert_score(score, old_min=300, old_max=850, new_min=250, new_max=900):
    # Scale the score from the old range to the new range
    old_range = (old_max - old_min)
    new_range = (new_max - new_min)
    new_score = (((score - old_min) * new_range) / old_range) + new_min
    return new_score

# Apply the conversion function
df_smes['CreditScoreConv'] = df_smes['CreditScore'].apply(convert_score)

# Drop column
df_smes.drop(['CreditScore'], axis=1, inplace=True)
df_smes.describe()
```

Detect Outliers

```python
#Indentify numerical for Outliers
df_num = df_smes.select_dtypes(include=['int64', 'float64'])
df_num.describe()

Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1
IQR
((df_num < (Q1 - 1.5 * IQR)) | (df_num > (Q3 + 1.5 * IQR))).any()
```

```python
#Boxplot outliers
fig = plt.figure(figsize=[32,24])
fig.suptitle('BOXPLOT OF NUMERICAL DATA', fontsize=25, fontweight='bold')
fig.subplots_adjust(top=0.92);
fig.subplots_adjust(hspace=0.5, wspace=0.4);
for i ,column in enumerate(df_smes.select_dtypes(include=['int64', 'float64'])):
    ax1 = fig.add_subplot(6,3, i+1);
    ax1 = sns.boxplot(data = df_smes, x=column);

    ax1.set_title(f'{column}')
    ax1.set_xlabel(f'{column}')
```

```python
from scipy import stats
```

```python
#Z-Score
print(f'Before filter outlier: {len(df_smes)}')

filtered_entries = np.array([True] * len(df_smes))
for col in [ 'Industry_Healthcare', 'Industry_Retail', 'Industry_Technology',
            'AnnualRevenueIDR', 'LatePaymentsIDR', 'CreditScoreConv'
            ]:
    zscore = abs(stats.zscore(df_smes[col]))
    filtered_entries = (zscore < 3) & filtered_entries

df_smes = df_smes[filtered_entries]

print(f'After filter outlier: {len(df_smes)}')
```

```python
# Drop Target Value 'CreditScoreConv'
predictor_features = ['Industry_Healthcare', 'Industry_Retail', 'Industry_Technology', 'AnnualRevenueIDR', 'LatePaymentsIDR']
#predictor_features = ['AnnualRevenueIDR', 'LatePaymentsIDR']


X = df_smes[predictor_features]
y = df_smes['CreditScoreConv']
```

```python
# Split Train - Test (Test size 30% rd6)
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.3, random_state=6)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, mean_absolute_error

reg_model = LinearRegression()
reg_model.fit(X_train, y_train)

y_pred_reg = reg_model.predict(X_val)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_val, y_pred_reg)
print("Mean Squared Error (MSE):", mse)

# Calculate Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_val, y_pred_reg, squared=False)
print("Root Mean Squared Error (RMSE):", rmse)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_val, y_pred_reg)
print("Mean Absolute Error (MAE):", mae)
```
